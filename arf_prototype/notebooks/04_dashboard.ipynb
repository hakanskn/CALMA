# 04 — Dashboard
Gradio panelini başlatır. `share=True` ile public ngrok URL üretir.
URL'yi tarayıcıda aç → tüm run'ları buradan yönet.

```python
import os, sys

REPO_URL = 'https://github.com/hakanskn/CALMA.git'
REPO_DIR = '/content/arf_prototype'

# Repo yoksa veya eksikse klonla
if not os.path.exists(os.path.join(REPO_DIR, 'dashboard')):
    os.system('rm -rf /content/CALMA')
    os.system(f'rm -rf {REPO_DIR}')
    ret = os.system(f'git clone {REPO_URL} /content/CALMA')
    if ret != 0:
        raise RuntimeError("git clone başarısız!")
    os.system(f'cp -r /content/CALMA/arf_prototype {REPO_DIR}')
    print('Repo hazır:', os.listdir(REPO_DIR))
else:
    print('Repo zaten mevcut.')

# Gradio 5.x — eski 4.31 sharing CDN artık çalışmıyor
os.system('pip install -q "gradio>=5.9.0" --upgrade')

if REPO_DIR not in sys.path:
    sys.path.insert(0, REPO_DIR)

os.environ.setdefault('DRIVE_BASE', '/content/drive/MyDrive/arf_results')
print('DRIVE_BASE:', os.environ['DRIVE_BASE'])
```
